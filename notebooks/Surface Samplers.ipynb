# Visualize surface grid

```python
cyl = Cyl(1, 1, toTransform([0,0,0],[0,0,0]))
```

```python
# Reset sampler each time
sampler = RegionSurfaceSampler(cyl, [130,40,30])

xs = []
ys = []
zs = []
num_samples = 5000

for i in range(num_samples):
    x,y,z = sampler.next() 
    xs.append(x)
    ys.append(y)
    zs.append(z)
    
```

```python
%matplotlib widget

# creating figure
fig = plt.figure()
ax = plt.axes(projection='3d')
  
# creating the plot
ax.scatter3D(xs, ys, zs, color=[0,0,1,0.1])
# plot_geeks = ax.scatter(xs_o, ys_o, zs_o, color=[1,0,0,0.1])
  
# setting title and labels
ax.set_title("Cylinder Sampling")
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
  
# displaying the plot
plt.show()
```
