```python
# import
from sklearn import datasets
import hypertools as hyp
import pickle as pkl
import matplotlib.pyplot as plt
```

```python
pickle_fn = '../data/region_0.25_0.5_2.pkl'
```

```python
%matplotlib widget
# %matplotlib inline

# Open a pickle file
with open(pickle_fn, 'rb') as f:
    data = pkl.load(f)

# Plot the data
hyp.plot(data, '.', ndims=3, reduce = 'TSNE')
```
